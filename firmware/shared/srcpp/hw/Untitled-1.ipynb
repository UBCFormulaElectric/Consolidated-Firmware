```python
from chimera_v2 import CRIT
import chimera_v2

c = CRIT()
```

```python
spi_led = chimera_v2.SpiDevice(c,"LED")
c.gpio_write("GPIO_LED_RCK", False)

spi_seven = chimera_v2.SpiDevice(c,"SEVEN_SEG")
c.gpio_write("GPIO_SEVEN_SEG_RCK", False)

```

```python
spi_seven.transmit(bytes([0xFF]))
spi_seven.transmit(bytes([0xFF]))
c.gpio_write("GPIO_SEVEN_SEG_RCK", True)
c.gpio_write("GPIO_SEVEN_SEG_RCK", False)
```

```python
while True:   
    spi_led.transmit(bytes([0xFF]))
    spi_led.transmit(bytes([0xFF]))
    c.gpio_write("GPIO_LED_RCK", True)
    c.gpio_write("GPIO_LED_RCK", False)
```

```python
c.gpio_write("GPIO_SEVEN_SEG_RCK", True)
c.gpio_write("GPIO_SEVEN_SEG_RCK", False)
```
